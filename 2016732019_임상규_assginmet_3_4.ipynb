### 4 실습 (Assignment 3)
* 행렬 방정식 풀기
- 다음 행렬 방정식을 'Pseudo inverse matrix'를 이용해 풀어보자
- $A^{T}A$의 역행렬이 존재한다고 가정

$$Ax=B$$

$$A = \begin{bmatrix}0 & 1 \\ 1 & 1 \\ 2 & 1 \\ 3 & 1 \end{bmatrix} $$

$$B = \begin{bmatrix}-1 \\ 0.2 \\ 0.9 \\ 2.1 \end{bmatrix} $$



```python
import numpy as np
import numpy.linalg as lin

matA = np.array([[0,1],[1,1],[2,1],[3,1]])
matA_ = lin.pinv(matA)
matB = np.array([-1,0.2,0.9,2.1])
print(np.dot(matA_, matB))
```

### 4 실습
* 2x2 그래프 그리기
- $y=x$
- $y=x^2$
- $y=sin(x)$
- $y=cos(x)$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)

plt.clf()
# 그래프 간격 조절
left=0.0
bottom=0.0
right=1.0
top=1.0
wspace=1.0
hspace=1.0
plt.subplots_adjust(left, bottom, right, top, wspace, hspace)
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.subplot(2,2,1)
plt.plot(x,x)
plt.title('x')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.subplot(2,2,2)
plt.plot(x,x*x)
plt.title('x^2')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.subplot(2,2,3)
plt.plot(x,np.sin(x))
plt.title('Sine')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.subplot(2,2,4)
plt.plot(x,np.cos(x))
plt.title('Cosine')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
